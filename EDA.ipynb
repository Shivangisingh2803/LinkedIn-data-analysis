```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('LinkedIn_DA_jobs_dataset.csv')
```

```python
print("Dataset Overview:-")
```

```python
print(data.info())
```

```python
print(data.head())
```

#handling missing values

```python
missing_value=data.isnull().sum()
```

#No missing values hence no need of handling missing values 

```python
print(missing_value)
```

```python
data.hist(figsize=(12, 10), bins=20)
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=data, y=col, order=data[col].value_counts().index)
    plt.title(f"Count Plot for {col}")
    plt.xlabel("Count")
    plt.ylabel(col.capitalize())
    plt.show()
```

```python
# Extract the city name by splitting the 'Location' column at the first comma
data['City'] = data['Location'].str.split(',').str[0]
```

```python
print(data['City'])
```

```python
data_cleaned = data.dropna()
```

```python
# Specify the file path and name where you want to save the file
output_file_path = "LinkedIn_DA_cleaned_data.xlsx"

# Save the cleaned DataFrame to an Excel file
data_cleaned.to_excel(output_file_path, index=False)

print(f"Cleaned data has been saved to {output_file_path}")
```
